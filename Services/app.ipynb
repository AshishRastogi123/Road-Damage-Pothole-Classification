```python
from flask import Flask, request, jsonify
from flask_cors import CORS
import tensorflow as tf
import numpy as np
from PIL import Image
from pyngrok import ngrok
from threading import Thread

app = Flask(__name__)

# ‚úÖ ENABLE CORS (MANDATORY)
CORS(app)

# üîê SET YOUR REAL NGROK TOKEN HERE
ngrok.set_auth_token(process.meta.env.NGROK_TOKEN)

# ‚úÖ Load model ONCE
model = tf.keras.models.load_model("/content/best_model.keras")

CLASS_NAMES = [
    "No Damage",
    "Minor Damage",
    "Moderate Damage",
    "Severe Damage",
    "Critical Damage"
]

@app.route("/predict", methods=["POST"])
def predict():
    if "image" not in request.files:
        return jsonify({"error": "No image provided"}), 400

    img = Image.open(request.files["image"]).convert("RGB").resize((224, 224))
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)

    preds = model.predict(img)[0]
    idx = int(np.argmax(preds))

    return jsonify({
        "predicted_class": CLASS_NAMES[idx],
        "confidence": float(preds[idx])
    })

# üîπ Run Flask on ONE port
def run_flask():
    app.run(port=5001)

Thread(target=run_flask).start()

# üîπ Expose SAME port via ngrok
public_tunnel = ngrok.connect(5001)

print("üöÄ Public URL:", public_tunnel.public_url)
print("üéØ Predict endpoint:", public_tunnel.public_url + "/predict")
```
